```python
import random
import numpy as np
```

```python
%timeit [random.random() for _ in range(100_000)]
```

```python
%timeit np.random.rand(100_000)
```

## Exercise 1 Conclusion                                                                                                                                                                                                                                                       
  NumPy was much faster than pure Python for generating 100,000 random numbers.
  
  - Python list comprehension: about 4.09 ms                                                                                                                 
  - NumPy: about 586 us                                                                                                                                      
  - NumPy was about 7x faster                                                                                                                                                                                            
  This happens because Python loops run in the interpreter, while NumPy uses optimized low-level code for array operations.
