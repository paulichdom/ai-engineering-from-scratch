## Exercise 1

```python
import random
import numpy as np
```

```python
%timeit [random.random() for _ in range(100_000)]
```

```python
%timeit np.random.rand(100_000)
```

### Exercise 1 Conclusion                                                                                                                                                                                                                                                       
  NumPy was much faster than pure Python for generating 100,000 random numbers.
  
  - Python list comprehension: about 4.09 ms                                                                                                                 
  - NumPy: about 586 us                                                                                                                                      
  - NumPy was about 7x faster                                                                                                                                                                                            
  This happens because Python loops run in the interpreter, while NumPy uses optimized low-level code for array operations.

## Exercise 2 - CSV Exploration

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    "model": ["Linear", "Random Forest", "Neural Net", "XGBoost"],
    "accuracy": [0.72, 0.89, 0.94, 0.91]
})
df
```

```python
df.to_csv("models.csv", index=False)
```

```python
df2 = pd.read_csv("models.csv")
df2
```

```python
plt.figure(figsize=(8,4))
plt.bar(df2["model"], df2["accuracy"])
plt.title("Model Accuracy")
plt.ylabel("Accuracy")
plt.xticks(rotation=20)
plt.show()
```

### Exercise 2 Conclusion                                                                                                                                     
In this exercise, I created a pandas dataframe, saved it as a CSV file, loaded it back from disk, and plotted the results with matplotlib.

This showed how Jupyter notebooks are useful for quick data exploration because code, tables, and charts appear in one place. It also showed that notebooks depend on running cells in the correct order, since imports must be available before using `pd` or `plt`.
